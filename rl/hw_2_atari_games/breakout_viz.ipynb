```python
import pickle

with open("rewards_list.pkl", "rb") as file:
    rewards_list = pickle.load(file)

with open("steps_list.pkl", "rb") as file:
    steps_list = pickle.load(file)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.lineplot(rewards_list, label="orig")
sns.lineplot(np.convolve(rewards_list, np.ones(10), "valid") / 10, label="avg_10")
plt.title("rewards")
```

```python
sns.lineplot(steps_list, label="orig")
sns.lineplot(np.convolve(steps_list, np.ones(10), "valid") / 10, label="avg_10")
plt.title("steps")
```
